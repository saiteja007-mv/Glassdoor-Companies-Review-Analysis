# Data Cleaning

```python
import pandas as pd
```

### Importing the RAW dataset for cleaning

```python
reviews_raw = pd.read_csv("C:\\Users\\saite\\OneDrive - University of Central Missouri\\Documents\\MS in CS\\Data Analyst\\Projects\\Employee Rentention Analysis\\Datasets\\Glassdoor Reviews Cleaned.csv")
```

```python
reviews_raw.head()
```

### Lookup if the dataset consists of any null values

```python
reviews_raw.isnull().sum()
```

#### Lets look how much percentage of data is missing in each column

```python
round(reviews_raw.isnull().mean() * 100,2)
```

```python
reviews_raw_copy = reviews_raw.copy()
```

#### Based on the percentage of missing data, the higher % of missing values are in Advice, RecommendtoFriend columns we will drop these columns which are not much important for analysis

```python
reviews_raw_copy.drop(columns=['Advice', 'rating RecommendtoFriend'],inplace=True)
```

#### For columns like ratingBusinessoutlook, ratingCEO we will keep these columns.
#### We will drop null values from columns like Employement Status, job title, location, Diversity and inclusion, Overall, SeniorLeadership, worklifebalance, summary

```python
reviews_raw_copy.dropna(subset=['Employement Status','job title','location','ratingDiversity and Inclusion', 'rating Overall', 'ratingSeniorleadership', 'rating worklifebalance', 'summary'],inplace=True)
```

```python
reviews_raw_copy.shape
```

#### So now the dataset shape reduced from (11925, 24) to (7160, 22) which is 39.96 ~ 40% of data has been removed.

```python
# How much percentage of data has been removed after cleaning
removed_data_percentage = round((1 - (reviews_raw_copy.shape[0] / reviews_raw.shape[0])) * 100, 2)
removed_data_percentage
```

## Now updating some columns data which can be used to analysis which are like date

```python
reviews_raw_copy['reviewdatetime'] = pd.to_datetime(reviews_raw_copy['reviewdatetime'])
```

## We will divide the datetime into two columns : Date and Time and then remove reviewDatetime column

```python
reviews_raw_copy['reviewdate'] = reviews_raw_copy['reviewdatetime'].dt.date
reviews_raw_copy['reviewtime'] = reviews_raw_copy['reviewdatetime'].dt.time
```

```python
reviews_raw_copy.drop(columns=['reviewdatetime'],inplace=True)
```

### Converting Data Types of ratingDiversity and Inclusion, rating Overall, ratingSeniorleadership, rating worklifebalance to int

```python
reviews_raw_copy['ratingDiversity and Inclusion'] = reviews_raw_copy['ratingDiversity and Inclusion'].astype(int)
reviews_raw_copy['rating Overall'] = reviews_raw_copy['rating Overall'].astype(int)
reviews_raw_copy['ratingSeniorleadership'] = reviews_raw_copy['ratingSeniorleadership'].astype(int)
reviews_raw_copy['rating worklifebalance'] = reviews_raw_copy['rating worklifebalance'].astype(int)
```

```python
reviews_raw_copy
```

### Splitting the location column into city and state and drop location column (Which can be used for visualization in PowerBI/Tableau)

```python
reviews_raw_copy['location'] = reviews_raw_copy['location'].str.split(',')
reviews_raw_copy['city'] = reviews_raw_copy['location'].str[0].str.strip()
reviews_raw_copy['state'] = reviews_raw_copy['location'].str[1].str.strip()
```

```python
reviews_raw_copy.drop(columns=['location'],inplace=True)
```

### Saving the changes to new CSV File

```python
reviews_raw_copy.to_csv("C:\\Users\\saite\\OneDrive - University of Central Missouri\\Documents\\MS in CS\\Data Analyst\\Projects\\Employee Rentention Analysis\\Datasets\\Glassdoor Reviews Cleaned and Processed.csv", index=False)
```
